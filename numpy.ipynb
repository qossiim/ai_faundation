```python
import numpy as np

arr = np.array([1,3,5,7])
print('numpy_array:', arr)
```

```python
arr2 = np.array([[1,3,5,7], [4,6,8,9]])
print('numpy_array2:', arr2)

```

```python
# dtype
a = np.array([1, 2, 3])
print(a.dtype)
```

```python
# reshape
a = np.array([1, 2, 3, 4, 5, 6 ])
b = a.reshape(2,3)
print("shaped_array:", b)
```

```python
# np.zeros()
a = np.zeros((2, 3))
print("zeros:\n", a)
```

```python
# np.ones()
a = np.ones((2, 3))
print("ones:\n", a)
```

```python
# np.full()
a = np.full((2, 3), 7)
print("full:\n", a)
```

```python
# np.arange()
a = np.arange(1, 11)
print("arange:", a)
```

```python
# np.linspace()
a = np.linspace(0, 10, 5)
print("linspace:", a)
```

```python
# np.random.rand()
a = np.random.rand(2, 3)
print("random:\n", a)
```

```python
# np.mean()
print("mean:", np.mean(a))
```

```python
# np.max()
print("max:", np.max(a))

# np.min()
print("min:", np.min(a))
```

```python
# np.sort()
a = np.array([5, 3, 8, 1, 2])
print("sort:", np.sort(a))
```

```python
# transpose()
a = np.array([[1, 2, 3],
              [4, 5, 6]])
print("transpose:\n", a.T)
```

```python
# flatten()
a = np.array([[1, 2, 3],
              [4, 5, 6]])
print("flatten:", a.flatten())
```
