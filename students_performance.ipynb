```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_performance.csv", on_bad_lines="skip")

print("Dataset shape (rows, columns):", df.shape)
print("Column names:", df.columns.tolist())

print("\nFirst 5 rows:")
print(df.head())

print("\nSummary statistics:")
print(df.describe())

print("\nData types and null info:")
df.info()
```

```python
print("\nMissing values:")
print(df.isnull().sum())
```

```python
print("\nAverage values:")
print(df[["weekly_self_study_hours", "attendance_percentage", "class_participation", "total_score"]].mean())
```

```python
print("\nMaximum values:")
print(df[["weekly_self_study_hours", "attendance_percentage", "class_participation", "total_score"]].max())
```

```python
print("\nMinimum values:")
print(df[["weekly_self_study_hours", "attendance_percentage", "class_participation", "total_score"]].min())
```

```python
print("\nMedian values:")
print(df[["weekly_self_study_hours", "attendance_percentage", "class_participation", "total_score"]].median())
```

```python
print("\nAverage total score by grade:")
print(df.groupby("grade")["total_score"].mean())
```

```python
print("\nStudent count by grade:")
print(df["grade"].value_counts())
```

```python
print("\nTop 10 students by total score:")
print(df.sort_values(by="total_score", ascending=False).head(10))
```

```python
print("\nStudents with score above 90:")
top_students = df[df["total_score"] >= 90]
print(len(top_students))
```

```python
print("\nStudents with score below 50:")
failed_students = df[df["total_score"] < 50]
print(len(failed_students))
```

```python
print("\nCorrelation Matrix:")
print(df[["weekly_self_study_hours", "attendance_percentage", "class_participation", "total_score"]].corr())
```
